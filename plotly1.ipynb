```python
import plotly.express as px
import numpy as np
```

```python
fig = px.line(x=np.arange(-4,4,0.1),y=np.sin(np.arange(-4,4,0.1)))
```

```python
fig.show()
```

```python
iris=px.data.iris()
```

```python
iris
```

```python
f1=px.line(iris,x='species',y='petal_width')
f1.show()
```

```python
f2=px.bar(iris,x='sepal_width',y='sepal_length')
f2.show()
```

```python
f3=px.histogram(iris,x='sepal_width',y='petal_width')
f3.show()
```

```python
f4=px.scatter(iris,x='species',y='petal_width')
f4.show()
```

```python
tips=px.data.tips()
```

```python
tips
```

```python
f5=px.pie(tips,values='tip',names='day')
f5.show()
```

```python
gap=px.data.gapminder()
```

```python
g1=px.scatter(gap,x='gdpPercap',y='lifeExp',size='pop',color='continent',log_x=True,
             animation_frame='year',animation_group='country',size_max=55,range_x=[100,100000],
             range_y=[25,90],hover_name='country')
g1.show()
```
