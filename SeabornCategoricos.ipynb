```python
import numpy as np
import scipy.stats as st
import seaborn as sb
import pandas as pd
```

```python
N0=st.norm.rvs(size=100000)
```

```python
sb.distplot(N0)
```

```python
st.describe(N0)
```

```python
N1=st.norm.rvs(5,1,size=100000)
```

```python
sb.distplot(N0)
sb.distplot(N1)
```

```python
B0=st.bernoulli.rvs(0.3,size=1000000)
B1=st.bernoulli.rvs(0.5,size=1000000)
B2=st.bernoulli.rvs(0.8,size=1000000)
```

```python
sb.distplot(B0)
sb.distplot(B1)
sb.distplot(B2)
```

```python
sb.histplot(B0)
sb.histplot(B1)
sb.histplot(B2)
```

```python
b0=st.binom.rvs(1000,0.3,size=1000000)
b1=st.binom.rvs(1000,0.5,size=1000000)
b2=st.binom.rvs(1000,0.8,size=1000000)
```

```python
sb.distplot(b0)
sb.distplot(b1)
sb.distplot(b2)
```
